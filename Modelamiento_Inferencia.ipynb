Importación de Librerias

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from time import time
from IPython.display import display
```

Consumimos los datos generados desde R Studio

```python
Data = pd.read_csv("Data_Limpia.csv")
Data
```

Eliminamos Los indices que heredamos del R studio

```python
Data = Data.drop(['Unnamed: 0'], axis=1)
Data
```

Partición del Data Frame en función de nuestras variables explicativas y variable objetivo

```python
y = Data.Fraude
x = Data.drop(['Fraude'], axis=1)
```

Previo al entrenamiento hice un escalamiento de las variables para buscar su normalización. Para ello utilicé la función  [MinMaxScaler](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html)

```python
scaler = MinMaxScaler()
# Para el uso de esta función es necesario hacer la conversion al tipo Array 
Data_Normalizada_array = scaler.fit_transform(x) 
columnas = list(Data.columns)
columnas.remove('Fraude')
Data_Normalizada_x = pd.DataFrame(Data_Normalizada_array,
    columns=columnas)
# Una muestra de los datos normalizados
display(Data_Normalizada_x.head(n = 5))
```

Para el entrenamiento hago uso de la función  [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) siguiendo el estandar 80/20 

```python
from sklearn.model_selection import train_test_split

# Defino las particiones para la data de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(Data_Normalizada_x, y, test_size = 0.2, random_state = 0)
print("Training set has {} samples.".format(X_train.shape[0]))
print("Testing set has {} samples.".format(X_test.shape[0]))
```

```python
from sklearn.metrics import accuracy_score, fbeta_score
def train_predict(learner, sample_size, X_train, y_train, X_test, y_test):
   '''
   inputs:
      - learner: the learning algorithm to be trained and predicted on
      - sample_size: the size of samples (number) to be drawn from training set
      - X_train: features training set
      - y_train: income training set
      - X_test: features testing set
      - y_test: income testing set
   '''
   # Creo un diccionario para los resultados
   results = {}

   # Entre y almaceno en learner
   start = time()
   learner = learner.fit(X_train[:sample_size], y_train[:sample_size])
   end = time()

   # Tiempo de entramiento fue:
   results['train_time'] = end - start

   # Predicciones
   start = time() 
   predictions_test = learner.predict(X_test)
   predictions_train = learner.predict(X_train[:300])
   end = time()

   # El tiempo total de las prediciones fue:
   results['pred_time'] = end -start

   # A continuación calculo la metrica acurracy en una muestra de la data de entrenamiento
   results['acc_train'] = accuracy_score(y_train[:300], predictions_train)

   # Ahora en la data de prueba
   results['acc_test'] = accuracy_score(y_test, predictions_test)

   # Repito le proceso pero con la metrica F para el entrenamiento
   results['f_train'] = fbeta_score(y_train[:300], predictions_train, beta=0.5) # Al ponerle 0.5 a beta, le estamos diciendo a la metrica que tenga prioridad sobre la precision del modelo

   # ahora aplico F, para Data de prueba
   results['f_test'] = fbeta_score(y_test, predictions_test, beta=0.5)

   # Guardo el modelo
   results['model'] = learner

   # retorno los resultados
   print("{} trained on {} samples.".format(learner.__class__.__name__, sample_size))  
   return results
```

```python
# LLevamos a acabo la ejecución y captura de métricas para la evaluación, todo ello almacenado en results.

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

random_state = 42

clf_RL = LogisticRegression(penalty='none')
clf_RF = RandomForestClassifier(random_state=random_state)
clf_GNB = GaussianNB()
clf_SVC = SVC(random_state=random_state)

samples = len(y_train)

results = {}
for clf in [clf_RL, clf_RF, clf_GNB, clf_SVC]:
    clf_name = clf.__class__.__name__
    results[clf_name] = train_predict(clf, samples, X_train, y_train, X_test, y_test)
```

```python
results
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer

# Lo realizado a continuación busca mejorar las metricas del modelo seleccionado mediante la modificación de parámetros.
clf = RandomForestClassifier(random_state = 42)
parametros = {}
parametros["n_estimators"]=[50,100,150]
parametros["max_depth"]=[10,20,30,40,50]
parametros["min_samples_leaf"]=[2,3,4,5]
parametros["min_samples_split"]=[2,4,8,10]
# 240 modelos ejecutados

Metrica_modelo=make_scorer(fbeta_score,beta=0.5)
Validacion_modelo = GridSearchCV(estimator=clf,param_grid=parametros,scoring=Metrica_modelo)
Validacion_modelo_entrenado=Validacion_modelo.fit(X_train,y_train)
Mejores_parametros=Validacion_modelo_entrenado.best_estimator_
```

```python
Predict_mejor_modelo=Mejores_parametros.predict(X_test)
fbeta_mejor_modelo=fbeta_score(y_test, Predict_mejor_modelo, beta=0.5)
```
